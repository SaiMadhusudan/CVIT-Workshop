### Problem 1

Given a list of binary strings `numbers`. Define a function that takes the first `k` strings and concats their parity into a string `encoding` and returns it.

````python
def getKstring(k,numberlist):
  if k > len(numberlist):
    return "K is larger than the numberlist"
  kstring = ""
  klist = numberlist[:k]
  for x in klist:
    kstring = kstring + str(x)
  
  return kstring

print(getKstring(3,["11","01","101","11"]))
````

**Solution**:

````python
#@title
def get_encoding(numbers,k):
  encoding = ""
  for index in range(k):
    parity = 0
    for digit in numbers[index]:
        parity ^= int(digit)
    encoding += str(parity)
  return encoding

print(get_encoding(["0","1","0","1"],3))
````

````python
fg = cv2.imread(f'drive/MyDrive/Workshop/images/fg.jpg')
fg = cv2.cvtColor(fg, cv2.COLOR_BGR2RGB)
bg = cv2.imread(f'drive/MyDrive/Workshop/images/bg.jpg')
bg = cv2.cvtColor(bg, cv2.COLOR_BGR2RGB)

def near_green(pixelvalue):
  if pixelvalue[1] > 200:
      return True
  return False

H=fg.shape[0]
W=fg.shape[1]

fg_copy = np.copy(fg)

for i in range(0,H):
  for j in range(0,W):
    if(near_green(fg[i][j])):
      fg_copy[i][j]=[0,0,0]

final_image = fg_copy + bg

plt.imshow(final_image)



````
